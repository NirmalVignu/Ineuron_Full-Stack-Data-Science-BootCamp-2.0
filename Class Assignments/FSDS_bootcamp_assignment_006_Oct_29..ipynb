### 1. write a function which will try to find out len of a string without using an inbuilt len function 

```python
# function to calculate length of string without using len()
def len_string(string):
    length=0
    for i in string:
        length+=1
    return length
len_string(input("Enter any string = "))
```

### 2. write a function which  will be able to print an index of list element without using an index function

```python
# function to print index of elemwnts in list without using index()
def index_element(lst):
    for i in range(len(lst)):
        print(f"{lst[i]} -> index= {i}")
index_element(['a','b',12,13,'INeuron'])
```

### 3. write a function which will be able to print an ip address of your system

```python
#print ipaddress of your system
import socket
def ip():
    hostname = socket.gethostname()
    ipaddrs = socket.gethostbyname(hostname)

    print("Your Computer Name is:" + hostname)
    print("Your Computer IP Address is:" + ipaddrs)
ip()
```

### 4. write a function which will shutdown your system

```python
# function to shutdown your system
import os,time
def shutdown():
    shutdown=input("Do you want to shutdown your computer? (y/n): ")
    if shutdown == "n":
        print("Shutdown failed")
    else:
        print("Shutdown initiated you have 10 sec for shutdown please save all the files and close all the programs ")
        time.sleep(10)
        print('Shutting down...')
        time.sleep(2) #here i'm using 3 sec sleep in order to save the changes to the file to show the above print statements
        os.system("shutdown /p /f")
shutdown()
```

### 5. write a function which will take input as a list with any kind of numeric value and give an out as a multiplication of all the numeric data l = [3.5, 6.56, 4,5,"sudh" , "ineuron" , 'fsda bootcamp 2.0']

```python
def numeric_multiplication(lst):
    mul=1
    for i in lst:
        if type(i)!=str:
            mul*=i
    return mul
print("numeric multiplication of [3.5, 6.56, 4,5,'sudh' , 'ineuron' , 'fsda bootcamp 2.0'] = ",numeric_multiplication([3.5, 6.56, 4,5,"sudh" , "ineuron" , 'fsda bootcamp 2.0']))
print("numeric multiplication of [1,2,3,4,5,6] = ",numeric_multiplication([1,2,3,4,5,6]))
```

### 8. write a function which will be able to read a doc/word file from your system

```python
import docx
def read_docx(file):
    doc = docx.Document(file)
    for i in doc.paragraphs:
        print(i.text)
read_docx('test.docx')
```
